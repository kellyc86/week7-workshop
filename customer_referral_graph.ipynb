```python
from raphtory import Graph
from raphtory import algorithms as algo
import pandas as pd

```

```python
# Create a new graph
graph = Graph()
```

```python
import random
# 2. List of 100 customers
NUM_CUSTOMERS = 100
customers = list(range(1, NUM_CUSTOMERS + 1))

# Add each customer as a node (timestamp=1 for simplicity)
for c_id in customers:
    graph.add_node(timestamp=1, id=str(c_id))

# 3. Randomly create referral edges
NUM_REFERRALS = 30  # or however many random edges you want
edges_created = set()  # to avoid duplicates (e.g., multiple edges A->B)

while len(edges_created) < NUM_REFERRALS:
    # Pick two distinct customers
    src = random.choice(customers)
    dst = random.choice(customers)
    if src != dst:
        # If not already in edges_created, add it
        edge_key = (src, dst)
        if edge_key not in edges_created:
            edges_created.add(edge_key)
```

```python
# 4. Add edges to the Raphtory graph
#    For each referral, we set a "timestamp" that can be anything (e.g., 2, 3,...).
timestamp_counter = 2
for (src, dst) in edges_created:
    graph.add_edge(
        timestamp=timestamp_counter,
        src=str(src),
        dst=str(dst),
        properties={
            "referral_source": str(src),  # or any property you want
            "referral_strength": random.choice([5, 10, 15, 20])  # sample property
        }
    )
    timestamp_counter += 1

```

```python
# 5. Inspect the graph summary
print(graph)
```

```python
# Example: PageRank
pr_result = algo.pagerank(graph)
top_list = pr_result.top_k(5)
print("Top 5 customers by PageRank:")
for node_id, score in top_list:
    print(f"Customer {node_id} — Score: {score}")
```

```python
# Convert to DataFrame
page_rank_df = pd.DataFrame(top_list, columns=["NodeID", "PageRankScore"])
page_rank_df
```

```python
cc_result = algo.weakly_connected_components(graph)
cc_list = cc_result.top_k(5)
print("Connected components (node -> component ID):", cc_list[:20], "...")
```

```python
# Get node names and degrees
most_active_nodes = [
    (node.name, node.degree()) for node in graph.nodes
]

# Sort by degree (descending)
most_active_nodes = sorted(most_active_nodes, key=lambda x: x[1], reverse=True)[:10]

# Print results
print("Top 10 Most Active Nodes:")
for node, degree in most_active_nodes:
    print(f"Node: {node}, Degree: {degree}")
```
